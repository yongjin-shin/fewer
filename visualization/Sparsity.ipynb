```python
from pruning.pruning_utils import plan_organizer
import matplotlib.pyplot as plt
```

```python
plan = [0,300,0]

reverse_plan = plan_organizer(plan, target_sparsity=0.9, plan_type='reverse', decay_type='linear')
server_plan = plan_organizer(plan, target_sparsity=0.9, plan_type='base', decay_type='linear')
```

```python
fig = plt.figure(figsize=(8,6))
plt.yticks([0.1+ 0.2*x for x in range(5)], fontsize=15)
plt.xticks([50*x for x in range(7)], fontsize=15)
plt.plot(reverse_plan, label='Fed-RP', lw=5, color='orange')
plt.plot(server_plan, label='Pruning', lw=5, color='blue')
plt.xlabel('Round', fontsize=20)
plt.ylabel('Sparsity', fontsize=20)
plt.xlim(0,300)
plt.ylim(0,1.0)
plt.legend(fontsize=18)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.grid()
```
